##3) What is the largest prime factor of the number 600851475143?

```python
def prime_factors(n):
    """Returns all the prime factors of a positive integer"""
    factors = []
    d = 2
    while n > 1:
        while n % d == 0:
            factors.append(d)
            n /= d
        d = d + 1
        if d*d > n:
            if n > 1: factors.append(n)
            break
    return factors


pfs = prime_factors(600851475143)
largest_prime_factor = max(pfs) # The largest element in the prime factor list
```

## Write a Python program to create a chain of function decorators (bold, italic, underline etc.)

```python
def make_bold(fn):
    def wrapped():
        return "</b>"+fn()+"</b>"
    return wrapped


def make_italic(fn):
    def wrapped():
        return '<i>'+fn()+'</i>'
    return wrapped

def make_underline(fn):
    def wrapped():
        return '<u>'+fn()+'</u>'
    return wrapped

@make_bold
@make_italic
@make_underline

def hello():
    return 'hello world'
print(hello())
```

```python
class NPRS:
    def __init__(self,data):
        self.data=data

    def total_votes(self,column_name_1,column_name_2):
        if isinstance(self.data, pd.DataFrame):
            df=self.data
            df['total_votes']=df[column_name_1]+df[column_name_2]
            return df
        else:
            try:
                df=pd.DataFrame(self.data)
                df['total_votes']=df[column_name_1]+df[column_name_2]
                return df
            except Exception as e:
                return print("Error converting data to DataFrame:", e)
                return None

    def pos_neg(self,column_name_1,column_name_2):
        df=self.total_votes(column_name_1,column_name_2)
        if df is not None:
            df['total_pos_less_neg'] = df[column_name_1] - df[column_name_2]
        return df

    def positive_prop(self,column_name_1,column_name_2,up_votes,total_votes):
        df=self.pos_neg(column_name_1,column_name_2)
        if df is not None:
            df['prop_pos'] = df[up_votes] / df[total_votes]
        return df

    def wilson_score(self,column_name_1,column_name_2,up_votes,total_votes,prop_pos,z):
        df=self.positive_prop(column_name_1,column_name_2,up_votes,total_votes)
        if df is not None:
            p_hat = df[prop_pos]
            n=df[total_votes]
            df['wilson_lower_bound'] = (p_hat + z**2/(2*n) - z*(((p_hat*(1-p_hat) + z**2/(4*n))/n)**0.5))/(1+z**2/n)
        return df

    def hacker_news_ranking(self,column_name_1,column_name_2,up_votes,total_votes,prop_pos,z,time,gravity):
        df=self.wilson_score(column_name_1,column_name_2,up_votes,total_votes,prop_pos,z)
        if df is not None:
            u=df[up_votes]
            d=df[column_name_2]
            t=time
            g=gravity
            df['hacker_score']=(u-d-1)/(t+2)**g
        return df



```

```python
# 1) function to check if the word is palindrome

def palindrome(string):
    reverse_string=string[::-1]
    if reverse_string==string:
        return f'the word {string} is palindrome'
    else:
        return f'the word {string} is not palindrome'
```

```python
# function to find the factorial of a number.

def factorial(num):
    if num == 0:
        return 1
    else:
        return num * factorial(num-1)



```

```python
def fac(num):
    if num == 0:
        return 1
    elif num < 0:
        raise ValueError('factorial not possible for negative numbers')
    result = 1
    for i in range(1, num+1):
        result *= i
    return result
```

```python
fac(4)
```

```python
# 7) functions to find if the number is a prime number

def prime(num):
    if num<2:
        return f'the number {num} is not prime'
    for i in range(2,num):
        if num%i==0:
            return f'the number {num} is not prime'
    return f'the number {num} is prime'
```

```python
prime(27)
```

```python
# 8) function to find common elements in from 2 lists

def common_elements(list1,list2):
    common=[]
    for i in list1:
        if i in list2:
            common.append(i)
    return f'List of common elements is {common}'
```

```python
# 9) sorting a list of elements in a list using bubble sort algorithm

def bubble_sort(arr):
    n = len(arr)  # Get the length of the array
    for i in range(n):  # Outer loop: iterate over the entire array
        swapped = False  # Initialize swapped to keep track of any swaps in this pass
        for j in range(0, n - i - 1):  # Inner loop: iterate from the start to the end of the unsorted portion
            # Compare adjacent elements
            if arr[j] > arr[j + 1]:
                # Swap if the element is greater than the next element
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True  # Set swapped to True if a swap occurred
        # If no swaps occurred, the array is already sorted
        if not swapped:
            break  # Exit the outer loop early if the list is sorted
    return arr


```

```python
bubble_sort([64, 34, 25, 12, 22, 11, 90])
```
